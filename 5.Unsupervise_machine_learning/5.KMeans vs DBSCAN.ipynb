# we need to import non linear data first

```python
from sklearnex import patch_sklearn
patch_sklearn()
```

```python
from sklearn.datasets import make_moons
```

```python
X,y = make_moons(
    n_samples=300,
    noise=0.05,
    random_state=42
        
)
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1])
```

# for this data KMeans algo

```python
from sklearn.cluster import KMeans
```

```python
kmean = KMeans(n_clusters=2)
```

```python
KMeans_label = kmean.fit_predict(X)
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],c=KMeans_label)
```

# NOW DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(
    eps=0.3,
    min_samples=4
)
```

```python
dbscan_label = dbscan.fit_predict(X)
```

```python
dbscan_label
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1], c=dbscan_label)
```

```python
# Summary 
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,6))

# First subplot
plt.subplot(1,3,1)
sns.scatterplot(x=X[:,0], y=X[:,1],c=y)
plt.title("ORIGINAL")

# Second subplot
plt.subplot(1,3,2)
sns.scatterplot(x=X[:,0], y=X[:,1], hue=KMeans_label, palette="deep")
plt.title("KMeans")

# Third subplot
plt.subplot(1,3,3)
sns.scatterplot(x=X[:,0], y=X[:,1], hue=dbscan_label, palette="deep")
plt.title("DBSCAN")

plt.tight_layout()
plt.show()
```
