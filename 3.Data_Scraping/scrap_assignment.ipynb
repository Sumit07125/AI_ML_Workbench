```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
with open("Scraped_data/data.html", "a") as f:
    for i in range(1, 101):
        stt = f"https://quotes.toscrape.com/page/{i}/"
        data = requests.get(stt)
        if data.status_code != 200:
            break
        f.write(data.text)
        f.write("\n\n")
```
