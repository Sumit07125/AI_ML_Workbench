## AI Summarizer Email and AI Assistant

```python
# queries ==> answer
# eamils => summerize
class PersonalAssistant:
    def __init__(self):
        print("Hi. I am AI Assistant , How can i Help you ?")

    def ans_query(self):
        question = input("Ask me anything :-)")

        
        response = client.responses.create(
        model="openai/gpt-oss-120b:free",
        max_output_tokens=500,
        temperature = 0.7,
        input=[{"role":"system" , "Content": "Act like a Help full personal assistant"},
              {"role":"user" , "content":question}]
        )
        print(response.output_text.strip())

    def summarize_email(self):
        email_text = input("Ask me anything :-)")
        prompt= f"summarize the following email in 2-3 sentences:{email_text}"
        

        
        response = client.responses.create(
        model="openai/gpt-oss-120b:free",
        max_output_tokens=500,
        temperature = 0.7,
        input=[{"role":"system" , "Content": "Act like a expert email Assistant"},
              {"role":"user" , "content":prompt}]
        )
        print("\n\n\nEmail Summary is : \n",response.output_text.strip())

        
```

```python
assistant = PersonalAssistant()
```

```python
assistant.ans_query()
```

```python
assistant.summarize_email()
```

## Image Genration

```python
import base64
```

```python
response = client.responses.create(
    model="openai/gpt-oss-120b:free",
    input="Generate an img of Highly Detailed Futuristic SpaceShip, Fly Near Nebula",
    tools=[{"type": "image_generation"}]
)

image_data = [
    output.result
    for output in response.output
    if output.type == "image_generation_call"
]

# decode and store img in file

if image_data:
    image_base64 = image_data[0]
    with open(r"spaceship.png", "wb") as f:
        image_bytes = base64.b64decode(image_base64)
        f.write(image_bytes)
```

#### Edit Img

```python
# EDIT

prompt = "Take the reference image and convert the spaceship color to red"

result = client.images.edit(
    model="gpt-image-1.5",
    prompt=prompt,
    image=[
        open("spaceship-image.png", "rb"),
    ]
)

image_base64 = result.data[0].b64_json
image_bytes = base64.b64decode(image_base64)

# Save the image to a file
with open("spaceship-red.png", "wb") as f:
    f.write(image_bytes)
```
